```python
import requests
import pandas as pd

# Replace 'YOUR_API_KEY' with your actual OpenFEC API key
API_KEY = 'BVTfalQokoMzRLDe1oCsI1k9T4OxyttZ56GQFOhZ'
ELECTION_YEAR = 2024  # Replace with your desired election year
PARTY = 'REP'  # R for Republican
OFFICE = 'H'
# &office={OFFICE} &election_year={ELECTION_YEAR}
# Base URL for the OpenFEC API request with parameters for active candidates
base_url = f'https://api.open.fec.gov/v1/candidates/?api_key={API_KEY}&party={PARTY}&candidate_status=C&election_year={ELECTION_YEAR}'

# Initialize an empty list to store all candidates
all_candidates = []
page = 1

while True:
    # Make a request to the OpenFEC API with the current page number
    response = requests.get(f'{base_url}&page={page}')
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        
        # Extract candidate information
        candidates = data.get('results', [])
        
        # If no candidates are returned, exit the loop
        if not candidates:
            break
            
        # Add candidates to the list
        all_candidates.extend(candidates)
        
        # Check pagination details
        pagination = data.get('pagination', {})
        total_pages = pagination.get('pages', 0)
        
        # If we're on the last page, exit the loop
        if page >= total_pages:
            break
            
        # Move to the next page
        page += 1
    else:
        print(f"Failed to retrieve data: {response.status_code}")
        break

# Create a Pandas DataFrame from the collected candidates data
df = pd.DataFrame(all_candidates)

# Display the DataFrame
print(df)
```

```python
wanted_columns = ['name', 'party', 'candidate_id', 'office', 'state', 'district']
df_active_republican_candidates = df[wanted_columns]
df_active_republican_candidates['name'] = df_active_republican_candidates['name'].str.replace('.', '', regex=False)

df_active_republican_candidates
```

```python
import random
# List of selected issues
issues = [
    "Healthcare",
    "Economy",
    "Education",
    "Environment",
    "Public Safety",
    "Immigration",
    "Social Justice",
    "Infrastructure",
    "Voting Rights",
    "Taxes"
]

# Function to generate a random mapping for issues
def generate_issue_map():
    return {issue: random.randint(1, 10) for issue in issues}

# Apply the function to create a new column with the issue mappings
df_active_republican_candidates['issue_map'] = df_active_republican_candidates.apply(lambda row: generate_issue_map(), axis=1)

# Display the updated DataFrame
df_active_republican_candidates
```
